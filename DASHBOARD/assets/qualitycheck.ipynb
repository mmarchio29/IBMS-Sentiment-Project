```python
keyword_count = df.groupby('Title')['Keyword'].nunique().reset_index()
keyword_count.columns = ['Title', 'Unique_Keywords']
```

```python
keyword_count.shape
```

```python
keyword_count.Unique_Keywords.value_counts()
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter

# Group by 'Title' and aggregate 'Keyword' into lists
grouped = df.groupby('Title')['Keyword'].apply(list).reset_index()

# Generate all unique pairs of keywords for each title
pair_counts = Counter()
for keywords in grouped['Keyword']:
    if len(keywords) > 1:
        pairs = combinations(sorted(keywords), 2)
        pair_counts.update(pairs)

# Convert the pair counts to a DataFrame
pair_counts_df = pd.DataFrame(pair_counts.items(), columns=['Keyword_Pair', 'Count'])

# Display the result
print(pair_counts_df)
```

```python
single_keyword_titles = keyword_count[keyword_count['Unique_Keywords'] == 1]

# Merge the original DataFrame with the filtered titles
filtered_df = df[df['Title'].isin(single_keyword_titles['Title'])]

# Display the result
filtered_df.head()
```

```python
filtered_df.Keyword.value_counts()
```

```python
grouped_filtered_df = filtered_df.groupby(['Keyword', 'Year']).mean().reset_index()
grouped_filtered_df.head()
```

```python
grouped_filtered_df.Keyword.value_counts()
```

```python
grouped_filtered_df['Sum'] = grouped_filtered_df.loc[:, 'Anger':'Surprise'].sum(axis=1)
grouped_filtered_df.head()
```

```python
grouped_filtered_df['Positive'] = grouped_filtered_df[['Happiness', 'Love', 'Surprise']].sum(axis=1)
grouped_filtered_df['Negative'] = grouped_filtered_df[['Anger', 'Disgust', 'Fear', 'Sadness']].sum(axis=1)
grouped_filtered_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(30, 8))
sns.lineplot(data=grouped_filtered_df.loc[grouped_filtered_df.Keyword == 'Communism'], x='Year', y='Positive', marker='o')
sns.lineplot(data=grouped_df.loc[grouped_df.Keyword == 'Communism'], x='Year', y='Positive')

plt.title('Positive Sentiment Trends from 1890 to 1960')
plt.xlabel('Year')
plt.ylabel('Sentiment Score')
plt.legend(title='Keyword')
plt.grid(True)
plt.show()
```
